1. Generating the first dataframe

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
df.drop(df[df.Borough == 'Not assigned'].index, inplace=True)
df.rename(columns = {'Postcode':'PostalCode'}, inplace = True)
df = df.groupby(['PostalCode','Borough'])['Neighbourhood'].apply(lambda x: ','.join(x)).reset_index()
# To check index of Not Assignedd df[df['Neighbourhood']=='Not assigned'].index
df.loc[85,'Neighbourhood'] = df.loc[85,'Borough']
print(df.shape)
df.head(12)
```

2. Generating 2nd Dataframe

```python
lat_lng_coords  = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lng_coords.rename(columns = {'Postal Code':'PostalCode'}, inplace = True)
df['latitude'] = lat_lng_coords['Latitude']
df['longitude'] = lat_lng_coords['Longitude']
df.head(12)
```

3. Find Coordinates of Toronto

```python
address = 'Toronto, CA'
geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

4. Generating Dataframe where Borough has Toronto

```python
df1 = df[df['Borough'].str.contains('Toronto')]
df1.reset_index(drop = True, inplace = True)
df1
```

```python
kclusters = 5

df_clustered = df1.drop(['Neighbourhood','Borough','PostalCode'], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df_clustered)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```
